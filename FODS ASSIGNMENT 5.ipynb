```python
from scipy import stats

# Function to remove outliers based on Z-score
def remove_outliers_zscore(group):
    z_scores = stats.zscore(group['Data Value'])
    abs_z_scores = abs(z_scores)
    filtered = group[(abs_z_scores < 3)]  # Keep data points with Z-score < 3
    return filtered

# Apply the Z-score outlier removal for each pollutant (Name)
df_clean_z = df.groupby('Name').apply(remove_outliers_zscore).reset_index(drop=True)

# Display the cleaned dataset
print(df_clean_z)
```

```python
cleaned_data=[]
name_of_pollutant=df['Name'].unique()

for pollutant in name_of_pollutant:
    pollutant_data=df[df['Name']==pollutant]

    Q1=pollutant_data["Data Value"].quantile(0.25)
    Q3=pollutant_data["Data Value"].quantile(0.75)
    IQR=Q3-Q1

    filtered_data=pollutant_data[(pollutant_data["Data Value"]>=Q1-1.5*IQR) & (pollutant_data["Data Value"]<=Q3+ 1.5*IQR)]
    cleaned_data.append(filtered_data)
df_clean=pd.concat(cleaned_data,ignore_index=False)
print(df_clean)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Name',y='Data Value',data=df_clean)
plt.title('Boxplot of data values with pollutant')
plt.xticks(rotation=90)
plt.show()
```

```python
#Sort data according to start date in ascending
df=df.sort_values(by='Start_Date')
df
```

```python
#Add a new column to original dataframe extracted from start_date column
df['year'] = pd.DatetimeIndex(df['Start_Date']).year
df
```

```python
#Year wise concentration of no2
no2_data=df[df["Name"]=="Nitrogen dioxide (NO2)"]
no2_data.groupby("year")['Data Value'].sum().plot(kind='line')
plt.title("Nitrogen concentration in air with respect to years")
plt.xlabel("year")
plt.ylabel("Data Value")
plt.xticks(ticks=no2_data['year'].unique(),rotation=90)
plt.show()
```

```python
#Average air quality value of pollutants in a particular area
plt.figure(figsize=(20,10))
df.groupby("Geo Place Name")["Data Value"].mean().sort_values().plot(kind='bar')
plt.title("Average value of air quality pollutants in a particular area")
plt.xlabel("Geo Place Name")
plt.ylabel("Data value")
plt.xticks(rotation=90)
plt.show()
```
