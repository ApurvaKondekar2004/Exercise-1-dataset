```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
mu, sigma = 100,4
data = np.random.normal(mu,sigma,100000)
```

```python
sns.histplot(data, bins=30, kde=True)
plt.title("Probability of height of men")
plt.xlabel("Height")
plt.ylabel("Probability")
plt.show()
```

```python
mu, sigma = 100, 4
data = np.random.normal(mu,sigma,100000)

data = (data - mu)/sigma

one_std = len(data[(data > -1) & (data < 1)])/len(data)*100
second_std = len(data[(data > -2) & (data < 2)])/len(data)*100
third_std = len(data[(data > -3) & (data < 3)])/len(data)*100

print("Percentage of data in one standard devaition",round(one_std,2))
print("Percentage of data in two standard devaition",round(second_std,2))
print("Percentage of data in three standard devaition",round(third_std,2))

sns.histplot(data, bins=30, kde=True)
plt.title("Standard normal ditribution of men's height")
plt.xlabel("Height")
plt.ylabel("Probability")
plt.show()
```

```python
#Using z table to find the area% under curve
from scipy.stats import norm
mu=100
sigma=4

z = (103 - mu)/sigma

p= norm.cdf(z)

print("Percentage of men shorter than 183 is",round(p*100,2))
```

```python
#Find the z-score from percentile
z = norm.ppf(0.6)
x = (z*sigma) + mu

print("The height for which 80% men are shorter:", round(x,2))
```
